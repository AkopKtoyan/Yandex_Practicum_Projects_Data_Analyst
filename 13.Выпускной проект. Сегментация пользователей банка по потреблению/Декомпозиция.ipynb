# Декомпозиция

## 1  Задача : Проанализировать клиентов регионального банка 'Метанпром' и сегментировать пользователей по количеству потребляемых продуктов, обращая особое внимание на вероятность оттока.

### 1.1  Определение целей исследования:

   Провести сегментацию пользователей по продуктам, которые им подходят, сегментацию по количеству потребляемых продуктов и признакам отточности и неотточности

   Результаты нашей работы могут быть интересны заказчику, а именно отделу маркетинга, менеджеры которого занимаются маркетинговыми исследованиями и формируют продуктовую линейку на основе клиентских предпочтений

   На основании проведенного анализа могут быть приняты решения о пересмотре продуктовой линейки, которая поможет удержать отточных клиентов и повысить лояльность среди неотточных клиентов

### 1.2  Ход исследования может содержать следующие этапы:

1) Загрузка данных, предобработка данных;

  *  исследовать пропущенные значения;

  *   исследовать соответствие типов;

  *  исследовать дубликаты;

  *  переименовать колонки;

  *  удалить дубликаты;

  *  привести типы;

  *  заменить пропущенные значения;

  *  кодирование категориальных признаков с помощью метода get_dummies

2) Проведение исследовательского анализа данных ;

  * для каждого признака построить гистограмму и boxplot, определить основные статистические показатели(среднее, моду, медиану, дисперсию);

  *  определить целевую переменную и выявить закономерности между целевой переменной и остальными признаками, используя сводную таблицу(pivot_table) или группировку(groupby);

  *  построение корреляционной матрицы (heatmap) для анализа взаимосвязи между признаками;

  *  построение матрицы диаграммы рассеяния(plotting.scatter_matrix) для определения степени взаимосвязи между признаками;

  *  категоризация признаков;

  *  подготовка данных к проверке статистических гипотез

3) Проверка статистических гипотез;

  *  формирование контрольной и экспериментальной выборок;

  *  выбор статистического теста и его обоснование;

  *  определение уровня статистической значимости;

  *  проверить гипотезу различия возраста между теми клиентами, которые пользуются 2 продуктами банка, и теми,которые пользуются одним;

  *  сделать выводы влияет ли возраст клиента на количество продуктов, которыми он пользуется;

  *  проверить гипотезу различия оценочного дохода между активными и неактивными клиентами;

  *  сделать выводы влияет ли оценочный доход клиента на то является ли он активным либо неактивным в последнее время

4) Сегментация пользователей на основе данных о количестве потребляемых продуктов и статусе активности или неактивности клиента;

  *   формирование более компактных сегментов на основе исследования отличительных черт групп;

  *   внутри каждого сегмента подсчитать процент ушедших;

  *   анализ портрета типового клиента;

5) Общие выводы;

    На основе проведенного исследования сформировать основные рекомендации для заказчика

6) Подготовка презентации и дашборда
