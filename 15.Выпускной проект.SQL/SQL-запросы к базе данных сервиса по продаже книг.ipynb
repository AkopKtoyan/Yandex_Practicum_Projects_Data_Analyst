2. Для каждой книги посчитайте количество обзоров и среднюю оценку

```python
# формируем запрос  
query = ''' SELECT title,
                   COUNT (DISTINCT review_id) AS review_count,
                   ROUND(AVG(rating), 2) AS avg_rating
            FROM books AS b 
            LEFT JOIN ratings AS r ON b.book_id = r.book_id 
            LEFT JOIN reviews AS rev ON b.book_id = rev.book_id
            GROUP BY b.book_id
            ORDER BY avg_rating DESC, review_count DESC '''
            
        

# выводим результат 
get_sql_data(query)


```

   В нашем распоряжении книги, отсортирвоанные по количеству обзоров и по среднему рейтингу. Книга с наибольшим средним рейтингом и количеством обзоров - это A Dirty Job (Grim Reaper #1) 

3. Определите издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры;

```python
# формируем запрос
query = ''' WITH publisher_count_books AS
            (SELECT publisher, 
                    COUNT(book_id) AS book_count 
             FROM publishers AS p 
             INNER JOIN books AS b ON p.publisher_id = b.publisher_id 
             WHERE num_pages > 50 
             GROUP BY publisher
             ORDER BY book_count DESC) 

             SELECT publisher
             FROM publisher_count_books
             WHERE book_count =  (SELECT MAX(book_count) 
                                 FROM publisher_count_books) '''
                                          
             
             
                      
# выводим результат                                     
get_sql_data(query) 
```

 Таким образом, 'Penguin Books' является издательством, которое выпустило наибольшее число книг толще 50 страниц

4. Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками

```python
# формируем запрос
query = ''' SELECT author, 
                   round(AVG(rating),2) AS avg_rating
            FROM authors AS a 
            JOIN books AS b ON a.author_id = b.author_id
            JOIN ratings AS r ON b.book_id = r.book_id
            WHERE b.book_id IN ( SELECT book_id 
                                FROM ratings 
                                GROUP BY book_id
                                HAVING COUNT(rating_id) >= 50 
                                ORDER BY COUNT(rating_id) DESC) 
            GROUP BY author 
            ORDER BY avg_rating DESC
            LIMIT 1 '''                    
            




# выводим результат 
get_sql_data(query)
```

Автором с самой выcокой средней оценкой книг (4.29) среди книг с 50 и более оценками - является J.K.rowling / Mary GrandPre

5. Посчитайте среднее количество обзоров от пользователей, которые поставили больше 48 оценок.

```python
# формируем запрос
query = '''  SELECT ROUND(AVG(review_cnt)) 
             FROM ( SELECT username, 
                           COUNT(review_id) AS review_cnt 
                    FROM reviews 
                    WHERE username IN  ( SELECT username 
                                         FROM ratings 
                                         GROUP BY username 
                                         HAVING COUNT(rating_id) > 48 ) 
             GROUP BY username 
             ) AS review_cnt  '''
                                     
            
            

# выводим результат 
get_sql_data(query)
```

Таким образом, среднее количество обзоров от пользователей, которые поставили больше 48 оценок, равно 24
